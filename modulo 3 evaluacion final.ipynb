```python
# 2.1 Elimina o trata los valores nulos, si los hay, en las columnas clave para asegurar que los datos estén completos

# # Comprobamos valores nulos en el DataFrame combinado
df_combinado.isnull().sum()
```

```python
# Definimos columnas clave a verificar

columnas_clave = ['Loyalty Number', 'Flights Booked', 'Total Flights', 'Salary' ,'Gender', 'Education', 'Province', 'Marital Status', 'Loyalty Card', 'Distance', 'Points Accumulated']
```

```python
# Imputamos los valores nulos en columna 'Salary'con la media

for column in ['Salary']:  # Ajusta según tus necesidades
    df_combinado[column].fillna(df_combinado[column].mean(), inplace=True)
```
