Fase 2: Visualización


Usando las herramientas de visualización contesta a las siguientes gráficas

```python
# 1. ¿Cómo se distribuye la cantidad de vuelos reservados por mes durante el año?

vuelos_por_mes = df_unido_copy.groupby(['Year', 'Month']).size().reset_index(name='Cantidad_vuelos')
vuelos_por_mes
```

```python
df_unido_copy['Year'].unique()
```

```python
# 2. ¿Existe una relación entre la distancia de los vuelos y los puntos acumulados por los cliente? Del resultado de la gráfica se pude conluir que la distancia de los vuelos no incide principalmente en los puntos acumulados

plt.figure(figsize=(8,6))
plt.scatter(df_unido_copy['Distance'], df_unido_copy['Points Accumulated'], color='blue')


plt.title('Relación entre distancia de vuelos y puntos acumulados')
plt.xlabel('Distancia del Vuelo')
plt.ylabel('Puntos Acumulados')
plt.grid()


plt.show()
```

```python
# 3. ¿Cuál es la distribución de los clientes por provincia o estado?

distribucion_por_provincia = df_unido_copy['Province'].value_counts().reset_index()

distribucion_por_provincia 
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Province', data=df_unido_copy, color = 'c', order=distribucion_por_provincia)


plt.title('Distribución de clientes por provincia')
plt.xlabel('Provincia')
plt.ylabel('Cantidad de clientes')
plt.xticks(rotation=45) 
plt.grid(axis='y')

plt.show()
```

```python
# 4. ¿Cómo se compara el salario promedio entre los diferentes niveles educativos de los clientes?

resultado_salario_promedio = df_unido_copy.groupby('Education')['Salary'].mean()
resultado_salario_promedio
```

```python
# 5. ¿Cuál es la proporción de clientes con diferentes tipos de tarjetas de fidelidad?

clientes_por_tipo_tarjeta = df_unido_copy['Loyalty Card'].value_counts()
clientes_por_tipo_tarjeta

proporcion_clientes = clientes_por_tipo_tarjeta / len(df_unido_copy)
proporcion_clientes
```
